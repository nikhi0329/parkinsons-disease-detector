```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Step 2: Simulate dataset (replace with real Parkinson’s dataset if you have one)
np.random.seed(42)
X = pd.DataFrame(np.random.rand(195, 22), columns=[f'feature{i}' for i in range(22)])
y = np.random.randint(0, 2, 195)

# Step 3: Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0, random_state=42)

# Step 4: Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 5: SVM model
model_svm = SVC()
model_svm.fit(X_train, y_train)
pred_svm = model_svm.predict(X_test)
acc_svm = accuracy_score(y_test, pred_svm)

# Step 6: Random Forest model
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
pred_rf = model_rf.predict(X_test)
acc_rf = accuracy_score(y_test, pred_rf)

# Step 7: Logistic Regression model
model_lr = LogisticRegression(max_iter=1000)
model_lr.fit(X_train, y_train)
pred_lr = model_lr.predict(X_test)
acc_lr = accuracy_score(y_test, pred_lr)

# Step 8: Print results
print(f"SVM Accuracy: {acc_svm:.4f}")
print(f"Random Forest Accuracy: {acc_rf:.4f}")
print(f"Logistic Regression Accuracy: {acc_lr:.4f}")

# Step 9: Plot accuracy comparison in specified order
algorithms = ["SVM", "Random Forest", "Logistic Regression"]
accuracies = [acc_svm, acc_rf, acc_lr]
#convert accuracies to percentages for diaplay on y-axis
accuracies_percentage=[acc*100 for acc in accuracies]
plt.figure(figsize=(8, 5))
plt.bar(algorithms, accuracies, color=["skyblue", "lightgreen", "orange"])
plt.ylim(0, 100)
plt.ylabel("Accuracy (%)")
plt.yticks(np.arange(0,101,10))
plt.title("Parkinson’s Disease Detection: Algorithm Comparison")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
